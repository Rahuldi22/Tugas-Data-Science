## RFM

In this project, we would segment the customers using customer's demography and purchasement behavior. To explain the customer's demography, we could use `Country` column while to explain the customer's purchasement behavior we have to transform our data first to get those information. We could use RFM analysis to get those information.

Dalam proyek ini, kami akan mengelompokkan pelanggan menggunakan demografi pelanggan dan perilaku pembelian. Untuk menjelaskan demografi pelanggan, kita dapat menggunakan kolom Negara, sedangkan untuk menjelaskan perilaku pembelian pelanggan kita harus mengubah data terlebih dahulu untuk mendapatkan informasi tersebut. Kita dapat menggunakan analisis RFM untuk mendapatkan informasi tersebut.

### Recency

To measure customer's recency, we could substract the latest transaction date in the data with each customer's latest transaction.

Untuk mengukur keterkinian pelanggan, kita dapat mengurangkan tanggal transaksi terakhir dalam data dengan transaksi terakhir setiap pelanggan.

```python
recency = df.groupby(['CustomerID'])['InvoiceDate'].max()
recency.head(10)
```

```python
max(recency)
```

Yang bagus adalah nilai recency yang kecil

```python
recency = max(recency) - recency
recency.head(10)
```

```python
# Get customer's recency
recency = df.groupby(['CustomerID'])['InvoiceDate'].max()
recency = max(recency) - recency
recency = recency.dt.days
recency = recency.rename('Recency')

recency.head()
```

```python
# Get customer's country of origin
country = (
    df
    .groupby(['CustomerID', 'Country_isUK'])
    .size()
    .reset_index()
    .drop(0, axis=1)
)

# Merge the data
df_new = (
    country
    .merge(recency, on='CustomerID')
    .merge(frequency, on='CustomerID')
    .merge(monetary, on='CustomerID')
)

# Set CustomerID as index
df_new = df_new.set_index('CustomerID')
df_new.head(20)
```

### One Hot Encoding

Since our model could just accept the numeric value, we have to recode the non numeric column.

Karena model kita hanya dapat menerima nilai numerik, kita harus mengkode ulang kolom non numerik.

```python
# Recode the non numeric variable

df_new = pd.get_dummies(
    df_new,
    columns=['Country_isUK'],
    drop_first=True,
    dtype=int

)

df_new = df_new.rename(
    columns={'Country_isUK_UK': 'Country_UK'}
)

df_new.head()
```

## Modeling

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# Modeling
possible_k = [2, 3, 4, 5, 6, 7, 8, 9, 10]
silhouette = []

for k in possible_k:
    kmeans = KMeans(n_clusters=k, random_state=42).fit(df_scaled)
    silhouette.append(silhouette_score(X=df_scaled, labels=kmeans.labels_))
```

```python
# Plot the score
plt.figure(figsize=(12, 7))
sns.lineplot(
    x=possible_k,
    y=silhouette,
    marker='*',
    markersize=10,
    markerfacecolor='red'
)
plt.xlabel('Values of K')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Method')
plt.grid()
plt.show();
```

Using silhouette score, the best K is either 2 or 6

Dengan menggunakan skor siluet, K terbaik adalah 2 atau 6

```python
kmeans = KMeans(n_clusters=6, random_state=42).fit(df_scaled)

df_result_kmeans = df_new.copy()
df_result_kmeans['cluster'] = kmeans.labels_

sns.pairplot(
    data=df_result_kmeans,
    hue='cluster',
    diag_kind='None',
    palette='tab10'
)

plt.show();
```

```python
# Quantile function
def q25(x):
    return x.quantile(0.25)

def q75(x):
    return x.quantile(0.75)

# Simple descriptive statistic
(
    df_result_kmeans
    .drop("Country_UK", axis = 1)
    .groupby('cluster')
    .aggregate(['mean', 'std', 'min', q25, 'median', q75, 'max'])
    .transpose()
)
```

```python
def rqs(recency_score):
  if recency_score < 17:
    return "High_R"
  elif 17 <= recency_score < 50:
    return "Medium_R"
  elif 50 <= recency_score < 141:
    return "Low_R"
  else:
    return "Bad_R"

df_result_kmeans["Recency_Quality"] = df_result_kmeans['Recency'].apply(rqs)
df_result_kmeans.head(20)
```

```python
df_result_kmeans["Recency_Quality"].value_counts()
```

```python
# recency_quatile_scoring: rqs
def rqs_if_else(recency_score):
  if recency_score < 17:
    return "High_R"
  elif 17 <= recency_score < 50:
    return "Medium_R"
  elif 50 <= recency_score < 141:
    return "Low_R"
  else:
    return "Bad_R"
```

```python
%timeit df_result_kmeans['Recency'].apply(rqs_if_else)
```

```python
import numpy as np

def rqs_np(recency_score):
    thresholds = [17, 50, 141]
    labels = ["High_R", "Medium_R", "Low_R", "Bad_R"]
    return labels[np.digitize(recency_score, thresholds, right=True)]
```

```python
%timeit df_result_kmeans['Recency'].apply(rqs_np)
```

# a. Bagaimana jika segmentation difokuskan untuk customer di luar UK? Apakah terdapat insight yang menarik?

```python
# Filter data pelanggan di luar UK
df_non_uk = df[df['Country'] != 'United Kingdom']

# Menghitung nilai total pembelian per CustomerID
total_purchase = df_non_uk.groupby('CustomerID')['TotalPrice'].sum()

# Menghitung frekuensi pembelian per CustomerID
frequency_purchase = df_non_uk.groupby('CustomerID')['InvoiceNo'].nunique()

# Menggabungkan total.purchase dan equency-purchase menjadi satu dataframe
customer_data = pd.concat([total_purchase, frequency_purchase], axis=1)
customer_data.columns = ['TotalPurchase', 'Frequency']
```

```python
# Normalisasi data
customer_data_normalized = (customer_data - customer_data.mean()) / customer_data

# Menggunakan KMeans untuk membuat 4 cluster
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(customer_data_normalized)

# Menambahkan label cluster ke dalam dataframe
customer_data['Cluster'] = kmeans.labels_

# Karakteristik/ciri utama dari masing-masing cluster non UK
cluster_characteristics = customer_data.groupby('Cluster').mean()
print(cluster_characteristics)

# Visualisasi karakteristik cluster
plt.figure(figsize=(10, 6))
sns.scatterplot(data=customer_data, x='Frequency', y='TotalPurchase', hue='Cluster')
plt.title('Karakteristik Cluster Pelanggan di Luar UK')
plt.xlabel('Fzekuensi Pembelian')
plt.ylabel('Total Pembelian')
plt.legend(title='Cluster')
plt.show()
```

# b. Apa karakteristik utama dari cluster-cluster non UK?

```python
from datetime import datetime

df_non_uk = df[df['Country'] != 'United Kingdom'].copy()

# Hitung nilai RFM
now = datetime.now()
df_non_uk.loc[:, 'Recency'] = (now - df_non_uk['InvoiceDate']).dt.days
df_non_uk.loc[:, 'Frequency'] = df_non_uk.groupby('CustomerID')['InvoiceNo'].transform('nunique')
df_non_uk.loc[:, 'Monetary'] = df_non_uk['Quantity'] * df_non_uk['UnitPrice']

# Normalisasi data
rfm_data = df_non_uk[['Recency', 'Frequency', 'Monetary']]
rfm_data = (rfm_data - rfm_data.mean()) / rfm_data.std()

# Terapkan K-Means
kmeans = KMeans(n_clusters=4, random_state=1)
df_non_uk.loc[:, 'Cluster'] = kmeans.fit_predict(rfm_data)

# Analisis karakteristik cluster
cluster_characteristics = df_non_uk.groupby('Cluster').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': ['mean', 'sum']
}).reset_index()

print(cluster_characteristics)
```
